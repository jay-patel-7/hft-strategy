```python
import yfinance as yf
import pandas as pd

# 1. Download data
# df = yf.download("^NSEI", start="1995-11-03")
df = yf.download("^NSEI", start="2010-01-14")

# 2. Reset index and prepare columns
df = df.reset_index()
df['Ticker'] = 'NIFTY 50'
df = df[['Ticker', 'Date', 'Open', 'High', 'Low', 'Close']]
df.columns = ['Ticker', 'date', 'open', 'high', 'low', 'close']

# 3. Reformat the date column to DD-MM-YYYY
df['date'] = df['date'].dt.strftime('%d-%m-%Y')

# 4. Save to Excel
df.to_csv("nifty_50_formatted.csv", index=False)

print("Data saved with formatted dates!")
```
