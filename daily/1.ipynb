```python
import yfinance as yf

# 1. Download data (ticker for NIFTY 50 is ^NSEI)
nifty_data = yf.download("^NSEI", start="2000-01-01", end="2026-04-05")

# 2. Add the Ticker column
nifty_data['Ticker'] = 'NIFTY 50'

# 3. Reset index to turn 'Date' into a column
nifty_data = nifty_data.reset_index()

# 4. Select and reorder the specific columns you requested
# Note: yfinance uses capitalized names by default, so we rename them
nifty_data = nifty_data[['Ticker', 'Date', 'Open', 'High', 'Low', 'Close']]

# Optional: Convert column names to lowercase if preferred
nifty_data.columns = ['Ticker', 'date', 'open', 'high', 'low', 'close']

print(nifty_data.head())
```

```python
import yfinance as yf
import pandas as pd

# 1. Download data
df = yf.download("^NSEI", start="2010-01-01")

# 2. Reset index and prepare columns
df = df.reset_index()
df['Ticker'] = 'NIFTY 50'
df = df[['Ticker', 'Date', 'Open', 'High', 'Low', 'Close']]
df.columns = ['Ticker', 'date', 'open', 'high', 'low', 'close']

# 3. Reformat the date column to DD-MM-YYYY
df['date'] = df['date'].dt.strftime('%d-%m-%Y')

# 4. Save to Excel
df.to_excel("nifty_50_formatted.xlsx", index=False)

print("Data saved with formatted dates!")
```
